```python
'''
You are given a DataFrame df with a single column sentence (string).

Your task is to extract all words from each sentence and return a new DataFrame with columns word and is_palindrome indicating whether the word is a palindrome.

Input DataFrame
Column Name	Data Type
sentence	String
Example Input
sentence
Madam Elle teaches malayalam
Wow, level up!
Output Schema
Column Name	Data Type
word	String
is_palindrome	Boolean
Example Output
word	is_palindrome
Madam	true
Elle	true
teaches	false
malayalam	true
Wow	true
level	true
up	false
Rules
A “word” is a contiguous sequence of alphabetic characters separated by space(ignore punctuation and numbers).
Palindrome check must be case-insensitive (e.g., Madam is a palindrome).
Keep duplicates (if a word appears multiple times across sentences, it should appear multiple times in output).
Starter Code
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

data = [
    ("Madam Elle teaches malayalam",),
    ("Wow, level up!",),
    ("No lemon, no melon",),
    ("Step on no pets.",),
    ("This is not a palindrome test",),
    ("Refer to civic duties at noon",),
    ("Racecar drivers drive fast",),
    ("Was it a car or a cat I saw?",),
    ("Stats and solos are fun",),
    ("Palindrome? Maybe, maybe not.",),
]

df = spark.createDataFrame(data, ["sentence"])

# Your logic goes here to create df_result with columns:
# - word (string)
# - is_palindrome (boolean)

display(df_result)
'''

# Initialize Spark session
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder.appName('Spark Playground').getOrCreate()

#Copy the starter code or load the file path available in the problem statement 
data = [
    ("Madam Elle teaches malayalam",),
    ("Wow, level up!",),
    ("No lemon, no melon",),
    ("Step on no pets.",),
    ("This is not a palindrome test",),
    ("Refer to civic duties at noon",),
    ("Racecar drivers drive fast",),
    ("Was it a car or a cat I saw?",),
    ("Stats and solos are fun",),
    ("Palindrome? Maybe, maybe not.",),
]

df = spark.createDataFrame(data, ["sentence"])

# keep only letters and spaces — replace everything else with space
df_clean = df.withColumn("clean", F.regexp_replace("sentence", "[^A-Za-z]+", " "))

df_result = (
  df_clean.withColumn("word", F.explode(F.split(F.col("clean"), "\\s+"))) # split into words
  .filter(F.col("word") != "") # remove blanks
  .withColumn("is_palindrome", F.lower("word") == F.reverse(F.lower("word"))) 
  .select("word", "is_palindrome")
)


# Display result.
df_result.show()
```
