```python
'''
You are given a dataset of product sales transactions. Each row represents a product, the store where it was sold, and the sales amount.

Your task is to find the Top N highest sales transactions per store. Consider N = 3 for solving the problem.

Input Schema
Column Name	Data Type
store	String
product	String
sales	Integer
Example Input Table
store	product	sales
StoreA	Laptop	800
StoreA	Mouse	25
StoreA	Monitor	300
StoreB	Keyboard	150
StoreB	Monitor	450
StoreB	Gaming PC	1200
StoreB	Printer	400
Output Schema
Column Name	Data Type
store	String
product	String
sales	Integer
If N = 2 â†’ Example Output Table
store	product	sales
StoreA	Laptop	800
StoreA	Monitor	300
StoreB	Gaming PC	1200
StoreB	Monitor	450
Starter Code
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

spark = SparkSession.builder.getOrCreate()

# Sample dataset
data = [
    ("StoreA", "Laptop", 800),
    ("StoreA", "Mouse", 25),
    ("StoreA", "Monitor", 300),
    ("StoreA", "Keyboard", 180),
    ("StoreA", "SSD", 90),
    ("StoreA", "Speakers", 250),
    ("StoreB", "Keyboard", 150),
    ("StoreB", "Monitor", 450),
    ("StoreB", "Gaming PC", 1200),
    ("StoreB", "Printer", 400),
    ("StoreC", "Macbook Air", 899),
    ("StoreC", "Air Pods", 399),
    ("StoreC", "Chromebook", 699),
    ("StoreC", "iPad", 699),
]
columns = ["store", "product", "sales"]
df = spark.createDataFrame(data, columns)

# Configurable N
N = 2

# Your code goes here
'''
# Initialize Spark session
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window

spark = SparkSession.builder.appName('Spark Playground').getOrCreate()

# Sample dataset
data = [
    ("StoreA", "Laptop", 800),
    ("StoreA", "Mouse", 25),
    ("StoreA", "Monitor", 300),
    ("StoreA", "Keyboard", 180),
    ("StoreA", "SSD", 90),
    ("StoreA", "Speakers", 250),
    ("StoreB", "Keyboard", 150),
    ("StoreB", "Monitor", 450),
    ("StoreB", "Gaming PC", 1200),
    ("StoreB", "Printer", 400),
    ("StoreC", "Macbook Air", 899),
    ("StoreC", "Air Pods", 399),
    ("StoreC", "Chromebook", 699),
    ("StoreC", "iPad", 699),
]
columns = ["store", "product", "sales"]
df = spark.createDataFrame(data, columns)

# Configurable N
N = 3

window_spec = Window.partitionBy("store").orderBy(F.col("sales").desc())

df_result = (
  df.withColumn("row_number", F.row_number().over(window_spec))
  .filter(F.col("row_number") <= N)
  .drop("row_number")
  .orderBy("store", F.col("sales").desc())
)

# Display result
df_result.show()
```
